```python
from sklearn.feature_extraction import DictVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
def train_and_evaluate(df_train1, df_val1, y_train1, y_val1, columns):
    """
    Trains a Logistic Regression model using the given columns and returns the accuracy score.
    """
    # Vectorize
    dv = DictVectorizer(sparse=False)
    train_dict = df_train1[columns].to_dict(orient='records')
    X_train = dv.fit_transform(train_dict)
    
    val_dict = df_val1[columns].to_dict(orient='records')
    X_val = dv.transform(val_dict)
    
    # Train model
    model = LogisticRegression(solver='liblinear', C=1.0, max_iter=1000, random_state=42)
    model.fit(X_train, y_train1)
    
    # Predict
    y_pred = model.predict_proba(X_val)[:, 1]
    y_pred_labels = (y_pred >= 0.5).astype(int)
    
    # Compute accuracy
    acc = accuracy_score(y_val1, y_pred_labels)
    
    return acc
```

```python
def feature_exclusion_experiment(df_train1, df_val1, y_train1, y_val1, all_columns):
    """
    Runs training with all columns and then with each column removed one at a time.
    Prints the resulting accuracy for each configuration.
    """
    # Run with all columns
    acc_all = train_and_evaluate(df_train1, df_val1, y_train1, y_val1, all_columns)
    print(f"all columns: {acc_all:.4f}")
    
    # Run with one column removed each time
    least_useful_col = ''
    least_useful_acc = 1
    least_useful_acc_delta = 1

    for col in all_columns:
        cols_subset = [c for c in all_columns if c != col]
        acc = train_and_evaluate(df_train1, df_val1, y_train1, y_val1, cols_subset)
        print(f"\nwithout '{col}': {acc:.4f}")
        cur_least_useful_acc_delta = abs(acc_all - acc)
        print("current usefulness:", cur_least_useful_acc_delta)
        if cur_least_useful_acc_delta < least_useful_acc_delta:
            least_useful_acc = acc
            least_useful_col = col
            least_useful_acc_delta = cur_least_useful_acc_delta

    print("\n\n")
    print("all_column_acc:", acc_all)
    print("least_useful_col:", least_useful_col)
    print("least_useful_acc:", least_useful_acc)
    print("Delta of accuracy with and without this column:", least_useful_acc_delta)
```

```python
from sklearn.feature_extraction import DictVectorizer
from sklearn.linear_model import LogisticRegression
import numpy as np

def accuracy_with_regularization(df_train1, df_val1, columns, y_train1, y_val1, C=1.0):
    """
    Train a Logistic Regression model with specified regularization (C)
    and return accuracy score on validation data.
    
    Parameters:
        df_train: pandas DataFrame - training data
        df_val: pandas DataFrame - validation data
        categorical: list[str] - categorical column names
        numerical: list[str] - numerical column names
        y_train: Series or array - training labels
        y_val: Series or array - validation labels
        C: float - inverse of regularization strength (default=1.0)
        
    Returns:
        float - validation accuracy
    """
    
    # Vectorize training data
    dv = DictVectorizer(sparse=False)
    train_dict = df_train1[columns].to_dict(orient='records')
    # print("Sample training record:\n", train_dict[0])
    
    X_train = dv.fit_transform(train_dict)
    # print("Feature names:\n", dv.get_feature_names_out())

    # Vectorize validation data
    val_dict = df_val1[columns].to_dict(orient='records')
    X_val = dv.transform(val_dict)

    # Train logistic regression model
    model = LogisticRegression(solver='liblinear', C=C, max_iter=1000, random_state=42)
    model.fit(X_train, y_train1)
    
    # print("Model coefficients:\n", model.coef_[0].round(3))
    # print("Intercept:", model.intercept_)
    
    # Predict probabilities for validation set
    y_pred = model.predict_proba(X_val)[:, 1]
    # print("Predicted probabilities:\n", np.round(y_pred[:10], 3))  # show first 10
    
    # Convert probabilities to binary predictions (using 0.5 threshold)
    will_convert = (y_pred >= 0.5)
    
    # Calculate accuracy
    accuracy_score = round((y_val1 == will_convert).mean(), 3)
    
    
    print(f"Validation Accuracy (C={C}): {accuracy_score:.4f}")
    return accuracy_score
```

```python
for c in [0.01, 0.1, 1, 10, 100]:
    accuracy_with_regularization(df_train, df_val, columns, y_train, y_val, C=c)
```
