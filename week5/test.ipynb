```python
import pickle
import numpy as np


def predict(datapoint):
    with open('pipeline_v1.bin', 'rb') as f:
        pipeline = pickle.load(f)
        print(type(pipeline))
        result = pipeline.predict_proba(datapoint)
        print(result)
        result = result[0,1]
        print(f'Probability of conversion: {result:.3f}')

```

```python
datapoint = {
    "lead_source": "paid_ads",
    "number_of_courses_viewed": 2,
    "annual_income": 79276.0
}
predict(datapoint)
```
